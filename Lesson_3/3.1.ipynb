```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.optimize import fsolve

def f(x, k, a, b):
    return (k * np.cos(x - a) + b)

x = np.linspace(-3*np.pi, 3*np.pi, 201)
plt.plot(x, f(x, 2, 10, 4))
plt.plot(x, f(x, 10, 5, 1))
plt.plot(x, f(x, 3, 7, 1))

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
