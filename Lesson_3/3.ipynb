```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x=1
y=0.5
a=2
b=1.5

t = np.linspace(0, 2*np.pi, 100)
plt.plot(x+a*np.cos(t) , y+b*np.sin(t))
```

```python
plt.figure(),
xlist = np.linspace(-np.pi, np.pi, 100)
ylist = np.linspace(-np.pi, np.pi, 100)
X,Y = np.meshgrid(xlist, ylist)
F = (X**2)/2.3 - (Y**2)/2 - 1
plt.contour(X, Y, F, [0])
```

```python
plt.figure()
xlist = np.linspace(-np.pi, np.pi, 100)
ylist = np.linspace(-2*np.pi/3, 2*np.pi/3, 100)
X,Y = np.meshgrid(xlist, ylist)
F = X**2 + Y**2 - 3
plt.contour(X, Y, F, [1])
plt.show()
```
